In this notebook we will develop the prediction model. In this case, EDA will not be significant since we have carried out and extensive data analysis in the previous notebooks, and the input for the model will be texts reviews since we are using Natural Language Processing (NLP) toolkits.

__Model Objetive:__ To classify a restaurant as top experience or non-top experience depending on customers text reviews.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm_notebook as tqdm
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, accuracy_score
from sklearn.ensemble import RandomForestClassifier
import networkx as nx
import shap
from sklearn.metrics.pairwise import cosine_similarity
import joblib # used but commented
from scipy.sparse import csr_matrix, save_npz # used but commented
```

```python
def evaluate_model(ytest, ypred, ypred_proba = None):
    if ypred_proba is not None:
        print('ROC-AUC score of the model: {}'.format(roc_auc_score(ytest, ypred_proba[:, 1])))
    print('Accuracy of the model: {}\n'.format(accuracy_score(ytest, ypred)))
    print('Classification report: \n{}\n'.format(classification_report(ytest, ypred)))
    print('Confusion matrix: \n{}\n'.format(confusion_matrix(ytest, ypred)))

def matriz_confusion(conf_matrix):
    sns.heatmap(conf_matrix, annot=True, fmt="d", linewidths=.5, square = True, cmap = 'Blues_r')
```

# Model Design.

```python
# We set the df of the food-related business in Pennsylvania merged with reviews, set in NB_2
foodpen_reviews = pd.read_csv('../data/Processed/foodpen_reviews_processed.csv')
foodpen_reviews.drop('Unnamed: 0',axis=1,inplace= True)
foodpen_reviews.head(10)
```

#### __Target Variable Selection__

In order to carry out a model using Natural Language Processing (NLP), we will group the original 5-values target variable in a new binary target variable. Therefore, we will categorise the reviews in whether the experience told in the reviews can be rated as Top experience (5 or 4 stars) or non-Top experience (1, 2 or 3 stars).

```python
foodpen_reviews['top_experience'] = foodpen_reviews['stars_review'].apply(lambda x: 1 if x >= 4 else 0)
foodpen_reviews.head(5)
    # foodpen_reviews.to_csv('../data/processed/foodpen_reviews_top_experience.csv')
```

```python
# We check the text of the reviews are clean enough
foodpen_reviews.text[85]
```

```python
# top experience distribution

labels = 'top_experience = YES', 'top_experience = NO'
sizes = foodpen_reviews["top_experience"].value_counts()
colors = ['lightpink', 'blue']
# Plot
plt.subplots(figsize=(8,8))
plt.pie(sizes, labels=labels, colors =colors, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

#### Text Vectorizing

In order to pass the text to the model we need to vectorise the text, so we transform it to weights to me parametrised in the model using sklearn feature extraction TfidVectorizer.

```python
%%time
# we are interested in the text of each review  and the stars rating, so we load these into separate list

texts = []
top_experience = [foodpen_reviews['top_experience'] for review in foodpen_reviews]
pbar = tqdm(total=foodpen_reviews.shape[0]+1)
for index, row in foodpen_reviews.iterrows():
    texts.append(row['text'])
    pbar.update(1)
pbar.close()
```

```python
%%time
from sklearn.feature_extraction.text import TfidfVectorizer

# This vectorizer breaks text into single words and bi-grams and then calculates the TF-IDF representation
vectorizer = TfidfVectorizer(ngram_range=(1,3))

# the 'fit' builds up the vocabulary from all the reviews while the 'transform' step turns each indivdual text into a matrix of numbers.
vectors = vectorizer.fit_transform(texts)
        # save_npz('../data/processed/vectors.npz', vectors)
```

```python
# We split into train and test with 85/15 proportion.
X_train, X_test, y_train, y_test = train_test_split(vectors, top_experience[1], test_size=0.35, random_state=None, shuffle=False)
```

# MODEL - SUPPORT VECTOR CLASSIFICATOR
        # https://medium.com/@bedigunjit/simple-guide-to-text-classification-nlp-using-svm-and-naive-bayes-with-python-421db3a72d34
        # https://www.kaggle.com/code/mehmetlaudatekman/text-classification-svm-explained

#### Training.

```python
%%time
classifier = LinearSVC()

# we train the model
model = classifier.fit(X_train, y_train)

# Saving the model
# joblib.dump(model, '../models/SVM.pkl')
```

#### Predictions.

```python
%%time
# Using the trained model to predict

predictions = model.predict(X_test)
print("REAL Ratings(Stars): " "")
display(y_test[:5])
print("PREDICTED Ratings: " "")
print(predictions[:5])
```

```python
%%time
# Predicting for entire dataset

X_null, X_full_test, y_null, y_full_test = train_test_split(vectors, top_experience[1], test_size=0.35, random_state=None, shuffle = False)
predict_all = model.predict(X_full_test)
```

```python
predicted_stars = list(predict_all)

print("Actual Ratings(Stars): ")
print(y_full_test[1050:1065])
print("\nPredicted Ratings: ","")
print(predicted_stars[1050:1065])
```

#### Results.

```python
conf_matrix = confusion_matrix(y_test, predictions)
print(classification_report(y_test, predictions))
matriz_confusion(conf_matrix)
print(f'model score on training data: {model.score(X_train, y_train)}')
print(f'model score on testing data: {model.score(X_test, y_test)}')
```

```python
TP = conf_matrix[0][0]
FP = conf_matrix[0][1]
FN = conf_matrix[1][0]
TN = conf_matrix[1][1]

total_correct = TP + TN
total_incorrect = FP + FN

TP_rate = TP / (TP + FN)
FP_rate = FP / (FP + TN)
TN_rate = TN / (TN + TP)
FN_rate = FN / (FN + TP)

success_ratio = (TP + TN) / (TP + FP + FN + TN)
failure_ratio = (FP + FN) / (TP + FP + FN + TN)

print('nº of predictions correctly estimated:', format(success_ratio, '0.2%'))
print('nº of predictions wrongly estimated:', format(failure_ratio, '0.2%'))
print('nº of top restaurants classified as non-top restaurants:', FN)
print('                percentage over total of top_restaurants',format(FN_rate, '0.2%'))
print('nº of non-top restaurants classified as top restaurants:', FP)
print('            percentage over total of non-top_restaurants', format(FP_rate, '0.2%'))
```

#### __Conclusion__:
We have obtained 0.92+ accuracy score for the model, which is pretty good considering that we are using texts for random users to estimate the outcome. In the confusion matrix, 9.29% of the real top_restaurants were classified as non-top_restaurants, meanwhile only 6.3% of the non-top restaurants(average or bad restaurants) were classified as top. Essentially, we can claim that the model is good-enough for our purpose.

# Interpretability

In order to use SHAP interpretability, we will use a random forest classifier, so we can see the outcome and the weights the model gives to every word used in the reviews, since we cannot use the SVC model to do it.

```python
foodpen_small = foodpen_reviews.iloc[::150]
X = foodpen_small['text']
y = foodpen_small['top_experience']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=7, stratify=y)
```

```python
vectorizer = TfidfVectorizer(min_df=10)
X_train_vec = vectorizer.fit_transform(X_train).toarray()
X_test_vec = vectorizer.transform(X_test).toarray()

model = RandomForestClassifier()
model.fit(X_train_vec, y_train)
```

```python
feature_names = vectorizer.get_feature_names_out()
explainer = shap.Explainer(model, X_train_vec, feature_names=feature_names)
shap_values = explainer(X_test_vec)
print(shap_values.values.shape)
```

```python
shap.initjs()
ind = 67

print(X_test.iloc[ind])

shap.plots.waterfall(shap_values[ind,:,1])
```

When we look at how the vectorised words in the text affects how the model predicts, we can see here (review nº67, chosen randomly) that there are words like 'minute', '20' or 'no' that help the model to choose the restaurant as non-top experience. Furthermore, if some words do not appear, like great o delicious, induce the model to categorise the experience as non-top as well. In this case, apparently there are no indicators of a top experience, so it can be easily deduced that the restaurant will not be granted as top-experience in the review.

```python
shap.initjs()
ind = 51

print(X_test.iloc[ind])

shap.plots.waterfall(shap_values[ind,:,1])
```

On the other hand, in this review we can see that there are multiple words that improves the rating of the reviews, making it to be classified as top experience. Even though the lexicon and the style of the review could be confused to the model, as the reviewer uses relatives verbal tenses, words like favorite, always, and specially perfect, induce the model to categorise it easily. The E[f(x)] represents the basis probability of, given a certain review, to be categorised as top-experienced, then, this probability is affected for the specific words from the review text to achieve a 0.85 final likelihood.

# Graphs

__What is a Graph (Data Structure)?__

A Graph is a non-linear data structure consisting of vertices and edges. The vertices are sometimes also referred to as nodes and the edges are lines or arcs that connect any two nodes in the graph. More formally a Graph is composed of a set of vertices( V ) and a set of edges( E ). The graph is denoted by G(E, V).

__Components of a Graph:__
- Vertices: Vertices are the fundamental units of the graph. Sometimes, vertices are also known as vertex or nodes. Every node/vertex can be labeled or unlabelled.
- Edges: Edges are drawn or used to connect two nodes of the graph. It can be ordered a pair of nodes in a directed graph. Edges can connect any two nodes in any possible way. There are no rules. Sometimes, edges are also known as arcs. Every edge can be labeled/unlabelled.

__Graphs in our dataset:__

In this case, we will use Graphs to see the relationship between the words of certain text samples from the customer reviews, so we can figure out how these words are related and hw a SVC model can take them into account and mmake predictions with them.

```python
foodpen_small = foodpen_small.reset_index(drop=True)
```

#### Example 1.

```python
# Vectorising text in reduced dataset
foodpen_sample = foodpen_small.iloc[0:1]
vectorizer = TfidfVectorizer()
small_vectors = vectorizer.fit_transform(foodpen_sample['text_clear'])
```

```python
similarity_scores = cosine_similarity(small_vectors)
```

```python

G = nx.Graph()

for i, row in foodpen_sample.iterrows():
    words = row['text_clear'].split()

    for word in words:
        G.add_node(word)

    for j in range(len(words) - 1):
        G.add_edge(words[j], words[j+1])

plt.figure(figsize=(12, 12))

nx.draw(G, pos=nx.spring_layout(G), with_labels=True)

for i, row in foodpen_sample.iterrows():
    print('Review:', row['text'], end='')
    plt.show()
```

In this first sample, which turns to be the first review of the dataset, we can see the conections between the word and how they are related in the text. If the words are located far from the center there's few relationship with the other words. Meanwhile, if they are close, it means they are closely related.
